### Data Cleansing and Descriptive Analysis

```python
def NA_Report(df):
    print('N/A Report:', '\n_______')
    [print(col, ': ', sum(df[col].isna()), sep='') for col in df.columns][0]
NA_Report(df)
```

```python
import warnings
warnings.filterwarnings("ignore")

def date_fix(df, col):
    for idx in range(len(df)):
        if int(df[col][idx][0:2]) > 12:
            date_split = df[col][idx].split('/')
            df[col].iloc[idx] = f'{date_split[1]}/{date_split[0]}/{date_split[2]}'
            
#date_fix(df_modeling, 'OrderDate')
date_fix(df, 'Order Date')
```

```python
def change_dtype(df):
    for col in df.columns:
        df[col] = df[col].astype(type(df[col][0])) 

#change_dtype(df_modeling)
```

```python
 #fillna with 0 (Not important feature and want to maximize level of data)
```

#### Pre-Process Analysis Data Frame

#### Pre-Process Modeling Data Frame

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(
    [('encoder', OneHotEncoder(), ['Segment', 'Country', 'Region', 'Category'])],
    remainder='passthrough'
)
#Nominal Encoding
encoded_data = ct.fit_transform(df)

#ohe.categories_[0]
```

```python
encoded_feature_names = ct.named_transformers_['encoder'].get_feature_names(['Segment', 'Country','Region', 'Category'])
encoded_feature_names
```

```python
df_modeling = pd.DataFrame()

df_modeling[['RowID', 'OrderID', 'OrderDate', 'ShipDate', 'ShipMode', 'CustomerID', 
            'CustomerName', 'City', 'State', 'PostalCode', 'ProductID', 
            'Sub-Category', 'ProductName', 'Sales']] = encoded_data[:, 11:]
df_modeling[encoded_feature_names] = encoded_data[:, 0:11]
```

```python
#Ordinal Encoding
df_modeling['ShipMode'] = df_modeling['ShipMode'].map({'Second Class': 0, 'Standard Class': 1,
                                                      'First Class': 2, 'Same Day': 3}) #Map Values
```

```python
df_modeling.head(2)
```

```python
change_dtype(df_modeling) #Change dtype of modeling
```

### EDA

```python
import matplotlib.pyplot as plt

# Group the dataframe and calculate the mean sales
grouped_sales = df.groupby(['Category', 'Sub-Category']).Sales.mean()

# Reset the index to convert the grouped series into a dataframe
grouped_sales = grouped_sales.reset_index()

# Pivot the dataframe to reshape it for the stacked bar chart
pivot_sales = grouped_sales.pivot(index='Category', columns='Sub-Category', values='Sales')

# Plot the stacked bar chart
pivot_sales.plot(kind='bar', stacked=True)

# Set the labels and title
plt.xlabel('Category')
plt.ylabel('Mean Sales')
plt.title('Mean Sales by Category and Sub-Category')

plt.legend(bbox_to_anchor=(1.04, 1), loc='upper left')

# Show the plot
plt.show()
```
