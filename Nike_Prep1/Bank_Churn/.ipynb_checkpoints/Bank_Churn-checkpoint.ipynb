### _Descriptive Analytics_ + _EDA_

```python
def numeric_col_avg(df):
    num_cols = df.select_dtypes(include=['float', 'int']).columns.tolist() #numeric columns
    for col in num_cols[2:]: #Avg for all numeric columns split on stayed/churned
        print(col)
        print(f"Stayed ({col}): {round(df[df['Exited'] == 0][col].mean(), 4)}")
        print(f"Churned ({col}): {round(df[df['Exited'] == 1][col].mean(),4)}") 
        print()

#numeric_col_avg(df)
```

#### _Churn Analysis Dashboard_

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig, axes = plt.subplots(3, 2, figsize=(37, 35))

#Plot Distributution of Balance
sns.histplot(x='Balance', data=df, hue='Exited', kde=True, ax=axes[0,0])
axes[0,0].set_title('Distribution of Balance');
# Data for the pie chart
complaints = [df.groupby('Complain').Exited.mean()[0], df.groupby('Complain').Exited.mean()[1]]
labels = ['Stayed', 'Churned']
colors = ['#66FF66', '#FF4C4C']  # Custom colors for the slices

# Create the pie chart
axes[0,1].pie(complaints, labels=labels, colors=colors, autopct='%1.1f%%')
axes[0,1].legend(labels, loc='upper right')
axes[0,1].set_title('Percentage of Complaints');

#Plot Age Difference
average_age_by_status = df.groupby('Exited')['Age'].mean().reset_index()

# Create the sideways bar chart # Set the figure size

sns.barplot(x='Age', y='Exited', data=average_age_by_status, orient='h', palette=['#66FF66', '#FF4C4C'],
           ax=axes[1,0])
axes[1,0].set_yticks([0, 1], ['Stayed', 'Exited'])
axes[1,0].set_xlabel('Average Age')
axes[1,0].set_ylabel('Status')
axes[1,0].set_title('Average Age of Employee Based on Churn');
fig.suptitle('Churn Analysis', fontsize=20);

#Plot correlation
sns.heatmap(round(df.drop(['RowNumber', 'CustomerId'], axis=1).corr(),2), annot=True, ax=axes[1,1])
axes[1,1].set_title('Correlation Matrix');

#Plot # Churned by Num. of Products w Reg line
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(np.arange(1,5).reshape(-1,1), df.groupby('NumOfProducts').Exited.sum().iloc[0:].values)
lr_model.coef_, lr_model.intercept_

lr_preds = []
for val in range(1,5):
    lr_preds.append(lr_model.predict([[val]])[0])

axes[2,0].plot(np.arange(1,5),df.groupby('NumOfProducts').Exited.sum().iloc[0:],
         marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='cyan', label='Avg. # Churned')
axes[2,0].plot(np.arange(1,5), list(map(lambda i: 0 if i<0 else i, lr_preds)), marker='s', c='r', ls= '--', lw=1.7,
        ms=5, mec='black', label='Regression')
axes[2,0].legend(loc='upper right')
axes[2,0].set_title('Avg. Churned v.s. Products held')
axes[2,0].set_xlabel('Number of Products')
axes[2,0].set_ylabel('Avg. Customers Churned')


# Plot Churned by Card Type
card_type_counts = df['Card Type'].value_counts()
# Plot the bar chart with gold color for the 'Gold' card type
card_type_counts.plot(kind='bar', color=['#808080', '#FFD700', '#C0C0C0', '#E5E4E2'], ax=axes[2,1])
axes[2,1].set_title('Churn by Card Type')
axes[2,1].set_xlabel('Card Type')
axes[2,1].set_ylabel('Avg. Customers Churned')
plt.savefig('ChurnAnalysis_Dashboard.pdf');
```

### _Decision Tree Classifier_

```python
from sklearn import tree
```

```python
df.drop(['Surname', 'CustomerId', 'RowNumber'], axis=1, inplace=True) #Remove unneccessary column
```

```python
df_modeling = pd.get_dummies(df, columns=['Geography', 'Gender', 'Card Type'])
df_modeling.head(3)
```

#### Split Features

```python
x = df_modeling.drop('Exited', axis=1)
y = df_modeling['Exited']

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.2, random_state=0)
```

#### Develop and Fit Model

```python
clf = tree.DecisionTreeClassifier()
clf.fit(x_train.values,y_train)
```

```python
y_train_pred = clf.predict(x_train.values)
y_test_pred = clf.predict(x_test.values)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train, y_train_pred))
print(accuracy_score(y_test, y_test_pred))
```

```python
#VISUALIZE THE TREE --> EXECUTE IN COLAB/DOWNLOAD GRAPHVIZ
import graphviz
feature_names = [str(feature) for feature in x.columns]
class_names = [str(cls) for cls in clf.classes_]

dot_data = tree.export_graphviz(clf, out_file=None, max_depth=5, feature_names=feature_names,
                               class_names=class_names, label='all', filled=True, special_characters=True)

graph = graphviz.Source(dot_data)
#graph
```

#### Feature importances
- Complaint - exceptionally strong predictor variable

```python
list(zip(clf.feature_importances_, x.columns))
```

#### GridSearch - hyperparameter optimization

```python
#search space
search_space = {
    'max_depth' : [5, 7, 9],
    'max_leaf_nodes': [6, 10, 14]
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

# Create the GridSearchCV object
gs= GridSearchCV(estimator=DecisionTreeClassifier(),
                           param_grid=search_space,
                           scoring='accuracy',
                           cv=5,
                           verbose=4)
```

```python
gs.fit(x_train, y_train) #FIT THE GRIDSEARCH
```

```python
gs.best_params_, gs.best_score_ #Best Parameters for Decision Tree Model
```

```python
# find average score/sd for the k-fold cross validation of each set/mix of parameters 
def print_results(gridsearch, metric=None):
    try:
        print('BEST PARAMS: {}\n'.format(gridsearch.best_params_))
        [print(f'{mean_test:.4f} +/-{std:.2f} for {params}')\
         for params, mean_test, std in zip(gridsearch.cv_results_['params'], \
                                           gridsearch.cv_results_['mean_test_score'], \
                                           gridsearch.cv_results_['std_test_score'])][0]
    except KeyError:
        print('BEST PARAMS: {} | {} Score: {}\n'.format(gridsearch.best_params_, metric, 
                                                        GS.cv_results_[f'mean_test_{metric}'].max()))
        [print(f'{mean_test:.2f} +/-{std:.2f} for {params}')\
         for params, mean_test, std in zip(gridsearch.cv_results_['params'], \
                                           gridsearch.cv_results_[f'mean_test_{metric}'], \
                                           gridsearch.cv_results_[f'std_test_{metric}'])][0]
        
print_results(gs)
```

### Random Forest Algorithm on Breast Cancer 

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
data = load_breast_cancer()
len(data.data), data.feature_names
```

#### _Feature Selection and Model Development_

```python
data = load_breast_cancer()
x = data.data
y = data.target.reshape(-1, 1)
data.target_names
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.25, random_state=0)

clf = DecisionTreeClassifier(random_state=0).fit(x_train, y_train)
clf
```

#### _Performance Evaluation_

```python
print ('Accuracy on Training Data =', accuracy_score(y_train, clf.predict(x_train)))
print ('Accuracy on Test Data =', accuracy_score(y_test, clf.predict(x_test))) #Overfit - Regularize!
```

#### Pruning to prevent overfitting (TREE #2)

```python
clf2 = DecisionTreeClassifier(max_depth=4, min_samples_leaf=8,
                             random_state=0).fit(x_train, y_train)
clf2
```

#### Performance Evaluation (TREE #2)

```python
y_pred = clf2.predict(x_test)
y_train_pred = clf2.predict(x_train)

print('Accuracy on Training Data =', accuracy_score(y_train, y_train_pred))
print('Accuracy on Testing Data =', accuracy_score(y_test, y_pred)) #Second Tree Accuracy
```

### Random Forest
- max_features k<m --> rule of thumb `k = sqrt(m)`
- n_estimators and max_features

```python
from sklearn.ensemble import RandomForestClassifier as FOREST
k = 8

my_forest = FOREST(max_features=k, random_state=0).fit(x_train, y_train.ravel())
```

```python
y_pred = my_forest.predict(x_test)
y_train_pred = my_forest.predict(x_train)

print('Accuracy on Training Data =', accuracy_score(y_train, y_train_pred))
print('Accuracy on Testing Data =', accuracy_score(y_test, y_pred)) 
```
