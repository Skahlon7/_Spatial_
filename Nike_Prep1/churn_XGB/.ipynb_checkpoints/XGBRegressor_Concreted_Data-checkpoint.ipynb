## XGB Regressor + GridSearchCV HyperParameter Optimization

#### Import Data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./concrete_data.csv')
df.head(3)
```

#### Feature Selection/Engineering

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,8].values.reshape(-1,1)
np.shape(x), np.shape(y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state=2021)
```

#### SearchSpace for Hyperparameters

```python
from xgboost import XGBRegressor
xgb_model = XGBRegressor(random_state=200) #Model Initilization

#Hyperparameter search dictionary
search_space = {
    'n_estimators' : [100,200,500],
    'max_depth' : [3,6,9],
    'gamma': [.01, .1],
    'learning_rate' : [.001, .01, .1, 1]
}
```

#### Initialize Gridsearch w/ Searchspace, CV, and Performance Metric

```python
from sklearn.model_selection import GridSearchCV
GS = GridSearchCV(estimator = xgb_model,
                 param_grid=search_space,
                 scoring = ['r2', 'neg_root_mean_squared_error'],#sklearn.metrics.SCORERS.keys() #view all
                refit='r2', #optimize on this scoring metric
                  cv=5, #5-fold Cross Validation
                  verbose= 4)
```

```python
GS.fit(x_train, y_train) #FIT THE GRIDSEARCH
```

#### EVALUATE GRIDSEARCH RESULTS

```python
GS.best_estimator_
```

```python
GS.best_params_, GS.best_score_
```

```python
GS.cv_results_['mean_test_r2'].max() #cv_results_ has all info of every model iteration
```

```python
# find average score/sd for the k-fold cross validation of each set/mix of parameters 
def print_results(gridsearch, metric):
    try:
        print('BEST PARAMS: {}\n'.format(gridsearch.best_params_))
        [print(f'{mean_test:.2f} +/-{std:.2f} for {params}')\
         for params, mean_test, std in zip(gridsearch.cv_results_['params'], \
                                           gridsearch.cv_results_['mean_test_score'], \
                                           gridsearch.cv_results_['std_test_score'])][0]
    except KeyError:
        print('BEST PARAMS: {} | {} Score: {}\n'.format(gridsearch.best_params_, metric, 
                                                        GS.cv_results_[f'mean_test_{metric}'].max()))
        [print(f'{mean_test:.2f} +/-{std:.2f} for {params}')\
         for params, mean_test, std in zip(gridsearch.cv_results_['params'], \
                                           gridsearch.cv_results_[f'mean_test_{metric}'], \
                                           gridsearch.cv_results_[f'std_test_{metric}'])][0]
```

```python
print_results(GS, 'r2')
```
