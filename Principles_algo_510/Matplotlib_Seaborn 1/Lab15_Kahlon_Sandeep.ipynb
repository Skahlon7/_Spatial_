You have to create line plots of the GDP(y-axis) against the year(x-axis) for 3 countries in Africa using Matplotlib in a single plot. 

*   The years chosen for this plot should be between 2010 and 2018(both inclusive). 
*   The countries chosen for this plot should be 'Equatorial Guinea', 'Mauritius'and 'Libya'.
*   marker='o' must be enabled for all the plots.
*   Legend must be present in the plot to show which line belongs to which conutry.
*   Label for x-axis must be 'Year', Label for y-axis must be 'Gross Domestic Product per capita (GDP)'.
*  Title of the plot must be 'GDP trend of three African countries'.

**Expected Output:**



download1.png

```python
#do not change this cell
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#write your code below
df = pd.read_csv('./gapminder_data_graphs.csv')
df_line = df.query('country.isin(["Equatorial Guinea", "Mauritius", "Libya"]) and year <=2018 and year >=2010 ')

plt.figure(figsize=(10,6))

for country in df_line.country.unique():
    plt.plot(df_line[df_line['country'] == country]['year'], df_line[df_line['country'] == country]['gdp'],
            marker = 'o')

plt.xlabel('Year'), plt.ylabel('Gross Domestic Product per capita (GDP)')
plt.legend([x for x in df_line.country.unique()])
plt.title('GDP trend of three African countries');
```

```python
#write your code below
#Filter DF
df_1998 = df.query('(continent=="Europe" or continent=="Africa") and year==1998')

fig, axes = plt.subplots(2, 2, figsize=(18, 10))

#plot1 1998 - scatter
axes[0,0].set_title('Life Expectancy (life_exp) vs % Service workers (services) - 1998')
sns.scatterplot(x=df_1998.services, y=df_1998.life_exp, hue=df_1998.continent, ax=axes[0,0])
axes[0,0].set_xlabel('% Service workers (services)')
axes[0,0].set_ylabel('Life Expectancy (life_exp)');

#Plot2 1998 - line
axes[1,0].set_title('GDP (gdp) vs % Service workers (services) - 1998')
sns.lineplot(x=df_1998.services, y=df_1998.gdp, hue = df_1998.continent, ax=axes[1,0])
axes[1,0].set_xlabel('% Service workers (services)')
axes[1,0].set_ylabel('GDP (gdp)');

#plot1 2018 - scatter
df_2018 = df.query('(continent=="Europe" or continent=="Africa") and year==2018')
axes[0,1].set_title('Life Expectancy (life_exp) vs % Service workers (services) - 2018')
sns.scatterplot(x=df_2018.services, y=df_2018.life_exp, hue=df_2018.continent, ax=axes[0,1])
axes[0,1].set_xlabel('% Service workers (services)')
axes[0,1].set_ylabel('Life Expectancy (life_exp)');

#plot2 2018 - line
axes[1,1].set_title('GDP (gdp) vs % Service workers (services) - 2018')
sns.lineplot(x=df_2018.services, y=df_2018.gdp, hue=df_2018.continent)
axes[1,1].set_xlabel('% Service workers (services)')
axes[1,1].set_ylabel('GDP (gdp)');
```
