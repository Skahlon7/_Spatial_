## Using re.search( ) like find( )

```python
hand = open('mbox.txt')
for line in hand:
    line = line.rstrip()  # Why?
    if line.find('From:') >= 0:
        print(line)
```

```python
hand = open('mbox.txt')
for line in hand:
    line = line.rstrip()
    if re.search('From:', line) :
        print(line)
```

## Using re.search( ) Like startswith( )

```python
hand = open('mbox.txt')
for line in hand:
    line = line.rstrip()
    if line.startswith('From:') :
        print(line)
```

```python
hand = open('mbox.txt')
for line in hand:
    line = line.rstrip()
    if re.search('^From:', line) :
        print(line)
```

```python
line = 'abcd +5.788 -32.32 45.4 fasdjfkl 22 0.55'
```

```python
re.findall('[+\-][0-9]+\.[0-9]+', line)
```

### what about possibly *unsigned* real numbers?

```python
re.findall('[+\-]{0,1}[0-9]+\.[0-9]+', line)
```

{m, n} indicagtes that the previous character should appear between m and n times

Obviously, there's more to regular expressions than we've covered here

A ? following a group means optional

```python
re.findall('[+\-]?[0-9]+\.[0-9]+', line)
```

### Greedy vs non-greedy matching

```python
# Greedy
x = 'From: Using the : character'
re.findall('^F.+:', x) 
```

```python
# Non-Greedy
re.findall('^F.+?:', x)
```

## Capture groups

Extract matched expression between parenthesis

```python
lin = 'From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008'
re.findall('^From .*@([^ ]*)', lin)
```

```python
m = re.match('^From .*@([^ ]*)', lin)
m
```

```python
m.group(0)
```

```python
m.group(1)
```

```python
lin = '''From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008
From louis@media.berkeley.edu Fri Jan  4 18:10:48 2008'''

re.findall(r'From .*@([\w.]*)', lin)

# re.findall('^From .*@([^ ]*)', lin)
```

```python
re.findall('From .*@([\w.]*)', lin)
```

```python
re.findall('^From (\S+@\S+)', lin)
```

### Matching across newlines

```python
paragraph = '''
<p>
This is a paragraph.
It has multiple lines.
</p>
'''
```

```python
re.search(r'<p>(.*)</p>', paragraph)
```

By default, the ‘.’ special character does not match newline characters :-|

But you can add a flag re.DOTALL to match them

```python
match = re.search(r'<p>(.*)</p>', paragraph, re.DOTALL)
match.group(0)
```

```python
match.group(1)
```

```python
re.findall(r'(.*)\n', paragraph)
```

### Maximum spam confidence

```python
import re
hand = open('mbox.txt')
numlist = list()
for line in hand:
    line = line.rstrip()
    stuff = re.findall('^X-DSPAM-Confidence: ([0-9.]+)', line)
    if len(stuff) != 1 :  continue
    num = float(stuff[0])
    numlist.append(num)
print('Maximum:', max(numlist))
```

```python
print(len(numlist))
print(numlist[:100])
```

### _findall_ with multiple capture group

```python
s = "The temperature was 20C (=68F). He drove a 1988 Ford Mustang."
re.findall(r'(\d+)\s*([FC])', s)  # two capture groups
```

```python
s = "The temperature was 20C (=68F). He drove a 1988 Ford Mustang."
re.findall(r'(\d+)\s*([FC])\b', s)  # boundary character after F or C
```

__\b__ marks a word boundary.

Careful with using normal strings, some escaped characters may not be properly processed. 
For example, \b is the backspace character in a regular string, but a word boundary in regualr expressions. Safer to use r'...' strings for REs.

```python
re.findall('(\d+)\s*([FC])', s)
```

```python
re.findall('(\d+)\s*([FC])\b', s) 
```

```python
re.findall(r'(\d+)\s*([FC])\b', s)
```

### Non-capture group

(?: …) establishes a unit without capturing towards m.group

```python
m = re.match(r'(-?\d+(?:\.\d+)?)', '98.6')
print(m)
m.group(0)
```

```python
m = re.match(r'(-?\d+(?:\.\d+)?)', '98')
print(m)
m.group(0)
```

### Backreferences \1 \2 ...

```python
s = '<table class="prop"> ... </table>'
m = re.match(r'<([a-z]+)\b[^>]*>.*?</\1>', s)
m.group(0)
```

```python
m.group(1)
```

```python
s = '<a href="https://www.isi.edu/"> ... </a>'
m = re.match(r'<([a-z]+)\b[^>]*>.*?</\1>', s)
print(m.group(0))
m.group(1)
```

```python
s = "3cats and 5dogs"
re.findall(r'([0-9])([a-z])',s)
```

```python
s = "3cats and 5dogs"
# \1 stands for the 1st captured group, \2 for the second captured group.
re.sub(r'([0-9])([a-z])', r'\1 \2', s)
```

### Positive and negative lookbehind, lookahead

```python
re.findall(r'-?\d+', '12-15')
```

```python
# Lookbehind: 
# (?<!\d)-?  match a - if not preceeded by a digit
re.findall(r'(?<!\d)-?\d+', '12-15')
```

```python
# replace "bar" with BAR if it is NOT preceded by "foo"
pattern = "(?<!foo)bar"
replacement = "BAR"
string = "foo bar foobar"

re.sub(pattern, replacement, string)
# 'foo BAR foobar'
```

### Compilation of Regular expressions

```python
def remove_special_characters(word):
    return re.sub('[^a-zA-Z ]', '', word)
```

```python
s = "I don't know her fiancé."
words = s.split()
print(words)

# map applies function to elements of list
print(list(map(remove_special_characters, words)))
```

```python
# compile a regular expression to make it faster
import re

regex_cc = re.compile('[^a-zA-Z ]')  # same with pre-compilation (advanced)
def remove_special_characters_compiled(word):
    return regex_cc.sub('', word)
```

```python
import timeit
```

```python
%timeit remove_special_characters("I don\'t know her fiancé.")
```

```python
%timeit remove_special_characters_compiled("I don\'t know her fiancé.")
```

## Regex Module

```python
import regex

spa_eng_dict = {"amigos": "friends", "hermana": "sister", 
                "mi": "my", "pequeña": "little", "tiene": "has"}

s = "María (mi hermana pequeña) tiene 1000 amigos 中国."
rest = s
result = ''

## r'(.*?)(\pL+)(.*)' 
#   .*?   match any chars non-greedy, 
#   \pL   followed one or more unicode code points in the "letter" category (i.e., a word)
#   .*    followed by any chars

while m3 := regex.match(r'(.*?)(\pL+)(.*)', rest):
    pre_word, word, rest = m3.group(1, 2, 3)
    print(pre_word, '|', word, '|', rest)
    # return translation or original word if no translation available
    result += pre_word + spa_eng_dict.get(word, word)
    print(result)
result += rest
result
```

### Lookahead


The (?=pattern) is known as positive lookahead. It performs the match, but does not capture the match, returning only the result: match or no match. It is also called assertion as it does not consume any characters in matching.

```python
regex = '^(?=.*?[A-Z])(?=.*?[a-z])(?=.*?\d)(?=.*?[\W_]).{8,}$'
password = 'My1#dfghj' # 'Mydfghj'
m = re.match(regex, password)
m.group(0)
```

```python
password = 'Mydfghj'
m = re.match(regex, password)
print(m)
```
