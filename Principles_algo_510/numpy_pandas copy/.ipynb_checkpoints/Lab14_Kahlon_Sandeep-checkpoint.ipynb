## **Question 2 - Mean Imputation[5 points]**

In the above dataset, you might have observed that the age column had a lot of blank/missing values. Inside the function 'impute' use the mean data imputation technique to fill in for missing values of the 'Age' column with the mean age of that column. Then calculate the mean age for the passengers by 'Sex' (male and female) and by 'Pclass' (1,2,3). The function should return a series with this tabulation of the mean ages (as shown below)

You can pass the same dataframe variable that you created in Question 1 to the impute() function.



```
Expected Output:

Pclass  Sex   
1       female    34.141405
        male      39.287717
2       female    28.748661
        male      30.653908
3       female    24.068493
        male      27.372153
```



**Hints:**

Check how fillna() works here - https://www.w3schools.com/python/pandas/ref_df_fillna.asp, you can use inplace parameter.

````python
def impute(df):
  #write your code below
    df.Age.fillna(df.Age.mean(), inplace=True)
    return df.groupby(['Pclass', 'Sex']).Age.mean()
````

````python
impute(df)
````
