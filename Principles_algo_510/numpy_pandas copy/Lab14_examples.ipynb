#### The np.array Method

#### Create a 2-dimensional Python list object, then convert it to a Numpy array object.

```python
ls = [[1, 2], [3,4], [5,6]]
arrNum = np.array(ls, dtype=object)
arrNum
```

#### ndarray Objects' Attributes

Play with the **ndim, shape, size, dtype, itemsize and data** attribute for the numpy array created above.


```python
print(arrNum.ndim)
print(arrNum.shape)
print(arrNum.size)
print(arrNum.dtype)
print(arrNum.itemsize)
print(arrNum.data)
```

#### Dimension of ndarray Ojects

Play with the reshape() and flatten() method.

```python
tempArr = arrNum.reshape(2,3)
print(tempArr)
print('\n')
print(arrNum.flatten())
```

#### The Slice Operation of ndarray Objects

Understand how the slice operation works for 1-D array and 2-D array.

```python
print(tempArr[1:])
print(tempArr[1:, 1:], '\n')

print(arrNum[1:])
```

#### The Calculation of ndarray Objects

Play with the **argmin(), argmax(), min(), max(), mean(), sum(), std(), dot(), square(), sqrt(), abs(). exp(), sign(), mod()** method.

```python
flat = arrNum.flatten()
np.max(flat)
np.min(flat)
np.std(flat)
```
