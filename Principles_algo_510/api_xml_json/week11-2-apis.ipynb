## API Example without API key: Currency conversion (JSON)

See https://github.com/fawazahmed0/currency-api#readme for API documentation

```python
import requests

from_currency = 'eur'
to_currency = 'usd'
base_url = 'https://cdn.jsdelivr.net/gh/fawazahmed0/currency-api@1/latest/currencies/'
url = base_url + f'{from_currency}/{to_currency}.json'
print(url)

r = requests.get(url)
j = r.json()

print(j)

print(f"1 Euro is {j.get('usd')} dollars")
```

## API Example without API key: NIH NCBI Entrez Pubmed papers (XML)

https://www.ncbi.nlm.nih.gov/home/develop/api/


```python
db = 'pubmed'
query = 'asthma[mesh]+AND+leukotrienes[mesh]+AND+2022[pdat]'

#assemble the esearch URL
base_url3 = 'https://eutils.ncbi.nlm.nih.gov/entrez/eutils/'
url3 = base_url3 + f'esearch.fcgi?db={db}&term={query}&usehistory=y'
url3
```

```python
r3 = requests.get(url3)
```

```python
r3.text
```

```python
from xml.etree import ElementTree

tree = ElementTree.fromstring(r3.text)
```

```python
key = tree.find('QueryKey').text
webenv = tree.find('WebEnv').text
print(key, webenv)
```

```python
url4 = base_url3 + \
f'efetch.fcgi?db={db}&query_key={key}&WebEnv={webenv}&rettype=abstract&retmode=xml'
url4
```

```python
r4 = requests.get(url4)
print(r4)
r4.text
```

### Aside: Pretty printing XML
https://www.geeksforgeeks.org/pretty-printing-xml-in-python/

```python
from bs4 import BeautifulSoup
  
soup = BeautifulSoup(r4.text, "xml")
print(soup.prettify())
```

```python
import xml.dom.minidom
  
xml = xml.dom.minidom.parseString(r4.content)

print(xml.toprettyxml() )
```

### Extracting Data from XML

```python
from lxml import etree

tree4 = etree.XML(r4.content)
tree4
```

```python
for title in tree4.xpath('//Article/ArticleTitle/text()')[:3] :
    print('-', title)
```

```python
for a in tree4.xpath('//Article')[:3] :
    print(a)
    print('-', a.find('ArticleTitle').text)
    # print('-', a.xpath('AuthorList/Author'))
    for au in a.xpath('AuthorList/Author'):
        print('  - ' , au.find('ForeName').text, au.find('LastName').text )
```

## API Example with API Key: Geocode

```python
import requests
import json
# api_key = '<top secret>'
with open('opencage.key') as f:
    api_key = f.read()
# print(api_key)

location_name = 'Los Angeles'
enc_loc = requests.utils.quote(location_name)
enc_loc
```

```python
base_url = 'https://api.opencagedata.com/geocode/v1/json'
url = f'{base_url}?q={enc_loc}&key={api_key}'
# url
```

```python
r = requests.get(url)
r.content
```

```python
j = r.json()
j
```

```python
results = j.get('results')
print(len(results))

places = [ x['formatted'] for x in results]

# places = list(map(lambda x : x['formatted'], results))

print('* Places called "Los Angeles":')
for p in places:
    print('-', p)
```

```python
la = results[0]
la
```

```python
la_osm_url = la['annotations']['OSM']['url']
la_osm_url
```

```python
print('lat:', la['annotations']['DMS']['lat'], ';  long:', la['annotations']['DMS']['lng'])
```
